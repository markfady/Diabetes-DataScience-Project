```python
#Splitting Data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X=dataset[dataset.columns[0:16]].values #inputs values
y=dataset[dataset.columns[17]].values   #outputs values
```

```python
# plot check for imbalanced data it takes more inputs but the used are what to write on x axis the name of dataset
# and the color (optional)
plt.figure(figsize=(8, 8))
dataset["class"].value_counts().plot.pie(autopct="%1.0f%%", colors=sns.color_palette("prism", 7), labels=["Positive", "Negative"],)
plt.title("balanced dataset check")
# splitting data to test and train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=1234)


```

```python
accuracies = {} #empty dictionary
from sklearn.metrics import recall_score
#Step 6 Creating Models 
#Logistic regression Model(1)
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
acc = logistic_regression.score(X_test, y_test)*100
accuracies['Logistic Regression'] = acc
y_pred = logistic_regression.predict(X_test)
# logistic regression confusion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
print('logistic regression Accuracy: ', metrics.accuracy_score(y_test, y_pred))
# calculate recall
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % recall)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
# random forest(3)
rf = RandomForestClassifier(n_estimators=2, random_state=1)
rf.fit(X_train, y_train)
acc = rf.score(X_test, y_test)*100
accuracies['Random Forest'] = acc
y_pred = rf.predict(X_test)
# random forest confusion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
print('Random Forest Accuracy: ', metrics.accuracy_score(y_test, y_pred))
plt.title("random forest confusion matrix")
# calculate recall
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % recall)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
# GaussianNB(4)
nb = GaussianNB()
nb.fit(X_train, y_train)
acc = nb.score(X_test, y_test)*100
accuracies['Naive Bayes'] = acc
y_pred = nb.predict(X_test)
# GaussianNB confusion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
print('GaussianNB Accuracy: ', metrics.accuracy_score(y_test, y_pred))
plt.title("GaussianNB")
# calculate recall
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % recall)
plt.show()
```

```python
from sklearn.svm import SVC
# svm(5)
svm = SVC(random_state=1, kernel='linear')
svm.fit(X_train, y_train)
acc = svm.score(X_test, y_test)*100
accuracies['SVM'] = acc
y_pred = svm.predict(X_test)
# svm confusion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
print('SVM Accuracy: ', metrics.accuracy_score(y_test, y_pred))
plt.title("svm confusion matrix")
# calculate recall
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % recall)
plt.show()
```
